*El archivo "Electric and Alternative Fuel Charging Stations" contiene información sobre estaciones de carga de vehículos eléctricos y de combustibles alternativos.*
*Este conjunto de datos es útil para aquellos interesados en encontrar estaciones de carga para vehículos que funcionan con energía eléctrica o combustibles alternativos, así como para análisis relacionados con la infraestructura de carga en diferentes áreas geográficas*.

**Cargamos Dataset**

```python
electric_alternative_fuel = pd.read_csv("../0-DataSets/0-DatosProporcionados/Electric and Alternative Fuel Charging Stations.csv")
```

```python
# Observamos las 5 primeras filas del dataset
electric_alternative_fuel.head()
```

```python
# Observamos la informacion que contiene el dataset
electric_alternative_fuel.info()
```

```python
# Descripcion del dataset
electric_alternative_fuel.describe()
```

```python
# Observamos las columnas que contiene el dataset
electric_alternative_fuel.columns
```

```python
# Revisamos cantidad de nulos
electric_alternative_fuel.isnull().sum()
```

```python
# Eliminamos las columnas que no vamos a utilizar en nuestro analisis
eliminar = ['Intersection Directions', 'ZIP', 'Plus4','Restricted Access',
       'Station Phone', 'Status Code', 'Expected Date',
       'Groups With Access Code', 'Access Days Time', 'Cards Accepted',
       'BD Blends', 'NG Fill Type Code', 'NG PSI', 'EV Level1 EVSE Num',
       'EV Level2 EVSE Num', 'EV DC Fast Count', 'EV Other Info', 'EV Network',
       'EV Network Web', 'Geocode Status','EV On-Site Renewable Source',
       'Date Last Confirmed', 'ID', 'Updated At', 'Owner Type Code',
       'Federal Agency ID', 'Federal Agency Name', 'Open Date',
       'Hydrogen Status Link', 'NG Vehicle Class', 'LPG Primary',
       'E85 Blender Pump', 'EV Connector Types', 'Country','LNG Vehicle Class',
       'Intersection Directions (French)', 'Access Days Time (French)',
       'BD Blends (French)', 'Groups With Access Code (French)',
       'Hydrogen Is Retail', 'Access Code', 'Access Detail Code',
       'Federal Agency Code', 'Facility Type', 'CNG Dispenser Num',
       'CNG On-Site Renewable Source', 'CNG Total Compression Capacity',
       'CNG Storage Capacity', 'LNG On-Site Renewable Source',
       'E85 Other Ethanol Blends', 'EV Pricing', 'EV Pricing (French)',
       'LPG Nozzle Types', 'Hydrogen Pressures', 'Hydrogen Standards',
       'CNG Fill Type Code', 'CNG PSI', 'CNG Vehicle Class']

electric_alternative_fuel.drop(eliminar, axis=1, inplace=True)
electric_alternative_fuel.head(2)
```

```python
electric_alternative_fuel = electric_alternative_fuel[electric_alternative_fuel['State'] == 'NY']
electric_alternative_fuel
```

```python
# Mantenemos solo las filas donde City = Manhattan, Brooklyn, Bronx, Queens, Staten Island
electric_alternative_fuel = electric_alternative_fuel[electric_alternative_fuel['City'].isin(['Manhattan', 'Brooklyn', 'Bronx', 'Queens', 'Staten Island'])]
electric_alternative_fuel
```

```python
# Renombramos las columnas
electric_alternative_fuel = electric_alternative_fuel.copy()
electric_alternative_fuel.rename(columns={'City':'Borough'}, inplace=True)
electric_alternative_fuel.head(2)
```

```python
# Eliminamos la columna State al no ser mas de utilidad
electric_alternative_fuel.drop('State', axis=1, inplace=True)
electric_alternative_fuel.head()
```

```python
# Buscamos filas repetidas
electric_alternative_fuel.duplicated().sum()
```

```python
# Eliminamos las filas repetidas
electric_alternative_fuel= electric_alternative_fuel.drop_duplicates()
```

**Guardamos el dataset en archivo Parquet**

```python
electric_alternative_fuel.to_parquet("../0-DataSets/2-DatosProporcionadosConETL/Electric and Alternative Fuel Charging Stations.parquet", index=False)
```

#### *Dataset Taxi_Zones*

*El archivo "taxi_zones" contiene información sobre las zonas de taxis, posiblemente en una ubicación específica, como una ciudad.*
*Este conjunto de datos probablemente se utiliza para mapear y administrar las zonas de taxis dentro de del estado de New York, lo que puede ser útil para la planificación y gestión del transporte público, así como para aplicaciones relacionadas con la movilidad urbana y la logística de transporte.*

**Cargamos el Dataset**

```python
taxi_zones = dbfread.DBF("../0-DataSets/0-DatosProporcionados/taxi_zones.dbf")
taxi_zones = pd.DataFrame(iter(taxi_zones))
taxi_zones
```

```python
# Observamos la informacion que contiene el dataset
taxi_zones.info()
```

```python
# Descripcion del dataset
taxi_zones.describe()
```

```python

# Observamos las 5 primeras filas del dataset
taxi_zones.head()
```

```python
# Eliminamos las columnas que no vamos a usar en nuestro análisis
eliminar = ['OBJECTID', 'Shape_Leng', 'Shape_Area']
taxi_zones.drop(eliminar, axis=1, inplace=True)
taxi_zones.head()
```

```python
# Renombramos las columnas
taxi_zones = taxi_zones.copy()
taxi_zones.rename(columns={'zone':'Zone', 'borough':'Borough'}, inplace=True)

taxi_zones.head()
```

**Guardamos el dataset en archivo parquet**

```python
taxi_zones.to_parquet("../0-DataSets/2-DatosProporcionadosConETL/taxi_zones.parquet", index=False)
```
