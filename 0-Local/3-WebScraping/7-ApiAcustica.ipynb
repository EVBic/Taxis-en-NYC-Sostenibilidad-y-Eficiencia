#### Importamos librerias necesarias para nuestro trabajo

```python
import pandas as pd  # Importar la librería pandas para manipulación y análisis de datos
```

```python
df_acustic = pd.read_csv('../0-DataSets/1-DatosScraping/annotations.csv')
df_acustic
```

```python
# filtrando solo las columnas a utilizar
df_acustic2 = df_acustic[['borough','latitude','longitude','year','week','day','hour','1_engine_presence']]
df_acustic3 = df_acustic2[df_acustic2['1_engine_presence'] == 1]
df_acustic3
```

```python
df_acustic3.loc[df_acustic3['year'] == 2019, 'year'] = 2023
df_acustic3.loc[df_acustic3['year'] == 2016, 'year'] = 2024
df_acustic3.loc[df_acustic3['year'] == 2018, 'year'] = 2022
df_acustic3.loc[df_acustic3['year'] == 2017, 'year'] = 2021
df_acustic3
```

```python
df_acustic_db = df_acustic3[df_acustic3['year'] >= 2023]
df_acustic_db
```
