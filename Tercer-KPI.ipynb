### 1. Segmentacion Geografica

### En este punto vamos a cargar los datasets para verificar como podemos medir el aumento de la segmentacion gegorafica de los taxis electricos

```python
import pandas as pd
```

```python
data_provista_estaciones = pd.read_csv(r'0-Local\0-DataSets\0-DatosProporcionados\Electric and Alternative Fuel Charging Stations.csv')
data_provista_estaciones.columns
```

```python
data_provista_estaciones = data_provista_estaciones.drop(columns=['Street Address', 'Intersection Directions', 'Station Phone', 'Groups With Access Code', 'Cards Accepted', 'BD Blends', 'NG Fill Type Code', 'NG PSI', 'EV Level1 EVSE Num','EV Level2 EVSE Num', 'EV DC Fast Count', 'EV Other Info', 'EV Network', 'EV Network Web', 'Date Last Confirmed','Updated At', 'Owner Type Code', 'Federal Agency ID', 'Federal Agency Name', 'Open Date', 'Hydrogen Status Link', 'NG Vehicle Class', 'LPG Primary', 'E85 Blender Pump', 'EV Connector Types','Intersection Directions (French)', 'Access Days Time (French)', 'BD Blends (French)', 'Groups With Access Code (French)', 'Hydrogen Is Retail', 'Access Code', 'Access Detail Code', 'Federal Agency Code', 'Facility Type', 'CNG Dispenser Num', 'CNG On-Site Renewable Source', 'CNG Total Compression Capacity', 'CNG Storage Capacity', 'LNG On-Site Renewable Source', 'E85 Other Ethanol Blends', 'EV Pricing (French)', 'LPG Nozzle Types', 'Hydrogen Pressures', 'Hydrogen Standards', 'CNG Fill Type Code', 'CNG PSI', 'CNG Vehicle Class', 'LNG Vehicle Class', 'EV On-Site Renewable Source', 'Restricted Access', 'Expected Date', 'Access Days Time', 'Plus4', 'EV Pricing', 'Geocode Status', 'ZIP'])
data_provista_estaciones
```

```python
data_provista_estaciones.dtypes
```

```python
data_provista_estaciones = data_provista_estaciones.loc[data_provista_estaciones['State'] == 'NY']
```

```python
data_provista_estaciones['City'].value_counts
```

```python
cities_to_keep = ['Bronx', 'Queens', 'Manhattan', 'Brooklyn', 'Staten Island']
data_provista_estaciones = data_provista_estaciones[data_provista_estaciones['City'].isin(cities_to_keep)]
```

```python
data_provista_estaciones['Fuel Type Code'].value_counts()
```

```python
data_provista_estaciones['Latitude'] = data_provista_estaciones['Latitude'].astype(str).str.replace('.', ',')
data_provista_estaciones['Longitude'] = data_provista_estaciones['Longitude'].astype(str).str.replace('.', ',')
```

```python
data_provista_estaciones
```

```python
data_provista_estaciones['Fuel Type Code'].value_counts()
```

```python
data_provista_estaciones['Fuel Type Code'] = data_provista_estaciones['Fuel Type Code'].replace(['BD', 'CNG', 'E85'], 'Fuel')
data_provista_estaciones
```

```python
data_provista_estaciones['Fuel Type Code'].value_counts()
```
